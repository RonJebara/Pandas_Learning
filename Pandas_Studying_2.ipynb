# Adding / Removing Rows

```python
import numpy as np
import pandas as pd
```

```python
bios = pd.read_csv('bios.csv')
bios_new = bios.copy()
bios_new.head()
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'])
bios_new
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new[(bios_new['born_year'] < 1975) & (bios_new['height_cm'] > 190)].sort_values('born_datetime', ascending=False)
```

```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
bios.head()
```

```python
bios[(bios['height_category'].isin(['Tall', 'Short'])) & (bios['born_country'].str.contains('FRA'))]
```

# Merging and Concatentating Data

```python
bios = pd.read_csv('bios.csv')
```

```python
nocs = pd.read_csv('noc_regions.csv')
nocs.head()
```

```python
bios_new = pd.merge(bios, nocs, left_on ='born_country', right_on ='NOC', how ='inner')
bios_new.head()
```

```python
bios_new.rename({'region': 'born_country_full'})
```

```python
bios_new[bios_new['NOC_x'] != bios_new['region']][['name', 'NOC_x', 'region']]
```

```python
usa = bios[bios['born_country'] == 'USA'].copy()
gbr = bios[bios['born_country'] == 'GBR'].copy()
```

```python
us_and_gbr = pd.concat([usa, gbr])
us_and_gbr.sample(n=15)
```

# shift(), rank(), rolling(), cumsum()

```python
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
coffee
```

```python
coffee.drop(columns = 'yesterday_revenue')
```

```python
coffee['change_in_revenue'] = coffee['revenue'] - coffee['revenue'].shift(2)
```

```python
coffee
```

```python
bios['height_rank'] = bios['height_cm'].rank(ascending = False)
```

```python
bios.sort_values(['height_rank'])
```

```python
coffee
```

```python
coffee['cumulative_revenue'] = coffee['revenue'].cumsum()
```

```python
coffee
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
```

```python
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
